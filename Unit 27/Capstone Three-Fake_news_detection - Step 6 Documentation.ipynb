# feature engineering

```python
## A. TOTAL NUMBER OF WORDS USED 
df_news['nb_words'] = df_news.text.apply(lambda x: len(x.split()))

## B. TOTAL NUMBER OF UNIQUE WORDS USED
df_news['nb_unique_words'] = df_news.text.apply(lambda x: len(set(x.split())))

## C. TOTAL NUMBER OF CHARACTERS USED
df_news['nb_char'] = df_news.text.apply(lambda x: len(x))

# D. TOTAL SPECIAL CHARACTERS USED

df_news["nb_special"] = df_news.apply(lambda p: sum( not q.isalpha() for q in p["text"] ), axis=1)

## D. TOTAL NUMBER OF PUNCTUATION USED
def punct(text):
    return(len([w for w in text.split() if w in list(string.punctuation)]))
df_news['nb_punct'] = df_news.text.apply(lambda x: punct(x))

## E. TOTAL NUMBER OF STOPWORDS USED
stopword = stopwords.words('english')
def stop(text):
    return(len([w for w in text.split() if w in stopword]))
df_news['nb_stopwords'] = df_news.text.apply(lambda x: stop(x))


## G. TOTAL NUMBER OF UPPERCASE WORDS USED
def upper(text):
    return(len([w for w in text.split() if w.isupper()]))
df_news['nb_title_case'] = df_news.text.apply(lambda x: upper(x))

## H. NUMBER OF MOST FREQUENT TERMS
token = nltk.word_tokenize(''.join(df_news.text))
frequent = nltk.FreqDist(token)
frequent.most_common(15)
```

```python
## I. NUMBER OF WORDS CONTAIN OUT OF MOST COMMON 100 WORDS 
freq_words = list(dict(frequent.most_common()[:100]).keys())
def freq(text):
    return(len([w for w in text.split() if w in freq_words]))
df_news['nb_freq_words'] = df_news.text.apply(lambda x: freq(x))
```

Most frequent words function:

```python
from nltk import tokenize

token_space = tokenize.WhitespaceTokenizer()
```

```python
import seaborn as sns
import nltk
    
def counter(text, column_text, quantity):
    all_words = ' '.join([text for text in text[column_text]])
    token_phrase = token_space.tokenize(all_words)
    frequency = nltk.FreqDist(token_phrase)
    df_frequency = pd.DataFrame({"Word": list(frequency.keys()),
                                   "Frequency": list(frequency.values())})
    df_frequency = df_frequency.nlargest(columns = "Frequency", n = quantity)
    plt.figure(figsize=(12,8))
    ax = sns.barplot(data = df_frequency, x = "Word", y = "Frequency", color = 'blue')
    ax.set(ylabel = "Count")
    plt.show()
```

# Building Vectorizer Classifiers

To get a good idea if the words and tokens in the articles had a significant impact on whether the news was fake or real, we should begin by using CountVectorizer and TfidfVectorizer.

```python
from sklearn.feature_extraction.text import CountVectorizer
# Initialize the 'count_vectorizer' 
count_vectorizer = CountVectorizer(stop_words='english')

# Fit and transform the training data 
count_train = count_vectorizer.fit_transform(X_train) 

# Transform the test set 
count_test = count_vectorizer.transform(X_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Initialize the 'tfidf_vectorizer' 
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7) 

# Fit and transform the training data 
tfidf_train = tfidf_vectorizer.fit_transform(X_train) 

# Transform the test set 
tfidf_test = tfidf_vectorizer.transform(X_test)
```
